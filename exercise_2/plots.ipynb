```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Size scalability

## Single precision

```python
ss = pd.read_csv("size-scalability-single.csv", sep=";")

ss
```

```python
ssblas = ss[ss["library"]=="openblas"]
ssmkl = ss[ss["library"]=="mkl"]
```

```python
fig, ax = plt.subplots()

ax.plot(ssblas["dim_M"],ssblas["GFLOPS"])
ax.plot(ssmkl["dim_M"],ssmkl["GFLOPS"])
ax.legend(["openblas","mkl"])
ax.set_title("Size scaling Single precision")
ax.set_xlabel("size")
ax.set_ylabel("GFLOPS")
```

## Comparison with Peak Performance

Theoretical Peak Performance
$$
FLOPS = \#cores \times Frequency \times IPC
$$

I used a single EPYC socket for this experiment. A CPU in EPYC has 64 cores, a frequency of 2.6 GHz and can perform 16 Floating Point double precision operations per cycle.
Theoretical Peak Performance for a single EPYC CPU is
$$
64 \times 2.6 \ GHz \times 16 = \boldsymbol{2.6624 \ TFLOPS}
$$

The best performance reached in the test is obtained by openBlas at the biggest dimension tested of 20'000 x 20'000 with $938.31 \ GFLOPS$, or $0.938 \ TFLOPS$; wich is only **35.24%** of the Theoretical Peak Performance.

## Different threads places

```python
sp = pd.read_csv("size-scalability-places.csv", sep=";")

sp
```

```python
spblas = sp[sp["library"]=="openblas"]
spmkl = sp[sp["library"]=="mkl"]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(7, 15))

ax1.plot(spblas[spblas["allocation policy"]=="threads"]["dim_M"],spblas[spblas["allocation policy"]=="threads"]["GFLOPS"])
ax1.plot(spmkl[spmkl["allocation policy"]=="threads"]["dim_M"],spmkl[spmkl["allocation policy"]=="threads"]["GFLOPS"])
ax1.legend(["openblas","mkl"])
ax1.set_title("Size scaling, Double precision, Threads allocation")
ax1.set_xlabel("size")
ax1.set_ylabel("GFLOPS")

ax2.plot(spblas[spblas["allocation policy"]=="cores"]["dim_M"],spblas[spblas["allocation policy"]=="cores"]["GFLOPS"])
ax2.plot(spmkl[spmkl["allocation policy"]=="cores"]["dim_M"],spmkl[spmkl["allocation policy"]=="cores"]["GFLOPS"])
ax2.legend(["openblas","mkl"])
ax2.set_title("Size scaling, Double precision, Cores allocation")
ax2.set_xlabel("size")
ax2.set_ylabel("GFLOPS")

ax3.plot(spblas[spblas["allocation policy"]=="sockets"]["dim_M"],spblas[spblas["allocation policy"]=="sockets"]["GFLOPS"])
ax3.plot(spmkl[spmkl["allocation policy"]=="sockets"]["dim_M"],spmkl[spmkl["allocation policy"]=="sockets"]["GFLOPS"])
ax3.legend(["openblas","mkl"])
ax3.set_title("Size scaling, Double precision, Sockets allocation")
ax3.set_xlabel("size")
ax3.set_ylabel("GFLOPS")
```
